```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# District Summary

    # Calculate the total number of schools
total_schools = len(school_data_complete.school_name.unique())
total_schools

    
    # Calculate the total number of students
student_total = school_data_complete ['student_name'].value_counts().sum()
student_total    
    
    # Calculate the total budget
budget_total = sum(school_data_complete.budget.unique())
budget_total   
    
    # Calculate the average math score 
avg_math_score = school_data_complete ['math_score'].mean()
avg_math_score

    # Calculate the average reading score
avg_reading_score = school_data_complete ['reading_score'].mean()
avg_reading_score
    
    # Calculate the percentage of students with a passing math score (70 or greater)
passing_math_score = (school_data_complete[school_data_complete['math_score'] >= 70].math_score.count()) / student_total
passing_math_score

    # Calculate the percentage of students with a passing reading score (70 or greater)
passing_reading_score = (school_data_complete[school_data_complete['reading_score'] >= 70].reading_score.count()) / student_total
passing_reading_score

    # Calculate the percentage of students who passed math and reading (% Overall Passing)
passing_m_r_score = (passing_math_score + passing_reading_score)/2

    # Create a dataframe to hold the above results
    
District_Summary = pd.DataFrame({"Total Schools":[total_schools],
                                 "Student Total":[student_total],
                                 "Budget Total":[budget_total],
                                 "Average Math Score":[avg_math_score],
                                 "Average Readying Score":[avg_reading_score],
                                 "% of Students Passing Math":[passing_math_score*100],
                                 "% of Students Passing Reading":[passing_reading_score*100],
                                 "% of Students Passing Math and Reading":[passing_m_r_score*100]   
    
    
})

District_Summary

    # Optional: give the displayed data cleaner formatting
```

```python
# School Summary
# Create an overview table that summarizes key metrics about each school, including:
school_summary = school_data_complete.groupby(['school_name'])
    # School Name
school_names = school_data_complete.school_name.sort_values().unique()
    # School Type
school_types = school_data.sort_values(by="school_name").type
    # Total Students
total_students = list(school_summary.student_name.count())
    # Total School Budget
total_budget = list(school_summary.budget.mean())
    # Per Student Budget
per_student_budget = [i/j for i,j in zip(total_budget,total_students)]
    # Average Math Score
avg_math_score = list(school_summary.math_score.mean())
    # Average Reading Score
avg_reading_score = list(school_summary.reading_score.mean())
    # % Passing Math
school_summary = school_data_complete[school_data_complete['math_score']>= 70].groupby(['school_name'])
pct_passing_math = [(i/j)*100 for i,j in zip(school_summary.math_score.count(),total_students)]
    # % Passing Reading
school_summary = school_data_complete[school_data_complete['reading_score']>= 70].groupby(['school_name'])
pct_passing_reading = [(i/j)*100 for i,j in zip(school_summary.reading_score.count(),total_students)]
    # % Overall Passing (The percentage of students that passed math and reading.)
school_overall_passing = [(i+j)/2 for i,j in zip(pct_passing_math,pct_passing_reading)]

    # Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({"School Names":school_names,
                                  "School Type":school_types,
                                  "Total Students":total_students,
                                  "Total Budget":total_budget,
                                  "Per Student Budget":per_student_budget,
                                  "Average Math Score":avg_math_score,
                                  "Average Reading Score":avg_reading_score,
                                  "% Passing Math":pct_passing_math,
                                  "% Passing Reading":pct_passing_reading,
                                  "Overall Passing Rate":school_overall_passing
    
})

school_summary_df = school_summary_df.reset_index(drop=True)
school_summary_df
```

```python
# Top Performing Schools (By % Overall Passing)
    # Sort and display the top five performing schools by % overall passing.

overall_top_5 = school_summary_df.sort_values(by="Overall Passing Rate", ascending=False).head(5)
overall_top_5
```
